```python
from google import genai
from google.colab import userdata

client = genai.Client(api_key=userdata.get('GOOGLE_API_KEY'))
```

```python
for m in client.models.list():
    print(m.name)
```

```python
model = "gemini-2.5-flash-lite" # @param ["gemini-2.5-flash-lite", "gemini-2.5-flash", "gemini-2.5-pro","gemini-3-pro-preview"] {"allow-input":true, isTemplate: true}
```

```python
from google.genai import types
```

```python
from IPython.display import Markdown
```

```python
# Define a 2025-standard configuration
my_config = types.GenerateContentConfig(
    # TEMPERATURE (0.0 to 2.0)
    # 0.0 = Precise/Factual | 1.0 = Balanced | 2.0 = Wildly Creative
    temperature=0.2,

    # THINKING CONFIG
    # We set a 'budget' of tokens the model can use just for 'internal reasoning'
    thinking_config=types.ThinkingConfig(
        include_thoughts=True,  # Let's see the "hidden" reasoning!
        thinking_budget=1024    # Max tokens for the thought process
    ),

    # CONSTRAINTS
    max_output_tokens=500,
    top_p=0.95
)
```

```python
# Call the model with our config
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain why 1+1=2 using Peano axioms.",
    config=my_config
)

# Print the reasoning (Thoughts) and the Answer
print("--- MODEL'S THOUGHT PROCESS ---")
for part in response.candidates[0].content.parts:
    if part.thought:
        print(part.text)

print("\n--- FINAL ANSWER ---")
display(Markdown(response.text))
```
